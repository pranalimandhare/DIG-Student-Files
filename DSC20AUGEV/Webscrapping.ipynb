```python

quotes=[]
for i in range(1,51):
  url = "http://quotes.toscrape.com/page/1/"
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')
  div = soup.find('div')
  divs = soup.find_all('div',class_="quote")

  for div in divs:
    title = div.find('span', class_='text' ).text
    small = div.find('small', class_='author' ).text
    tags = div.find('a', class_='tag' ).text
    quotes.append([title, small, tags])
print(quotes)
```

```python
df = pd.DataFrame(quotes, columns = ['title', 'Author', 'Tags'])
df
```

```python
books=[]
for i in range(1,51):
  url = "https://quotes.toscrape.com/page/1.html"
  response = requests.get(url)
  soup = BeautifulSoup(response.content, 'html.parser')
  ol = soup.find('ol')
  articles = ol.find_all('article', class_='product_pod' )

  for article in articles:
    image = article.find('img')
    title = image.attrs['alt']
    star = article.find('p')
    star = star['class'][1]
    price = article.find('p', class_ = 'price_color').text
    price = float(price[1:])
    books.append([title, star, price])
```
